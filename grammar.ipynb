```python

#提取出来的值有list和dict两种类型，分别处理 
#   
#getName函数递归遍历字典，删除无用的键值对，只保留'name''members'和'content'
def getName(dictionary):
    if type(dictionary)==dict:
        for key in list(dictionary.keys()):
            if key!='name'and key!='members' and key!='content':
                dictionary.pop(key)
            elif key=='members':
                getName(dictionary['members'])
            elif key=='content':
                getName(dictionary['content'])
    elif type(dictionary)==list:
        for member in dictionary:
            getName(member)



#structAdjust函数调整字典结构，把members和content以及name等关键词去除并修改合适结构
def structAdjust(dictionary):
    if type(dictionary)==dict:
        if 'members' in dictionary.keys():
            if 'name' in dictionary.keys():
                members=[]
                for member in dictionary['members']:
                    #print('member:')
                    #print(member)
                    #print('\n')
                    members.append(structAdjust(member))
                return {dictionary['name']:members}
                
            else:
                members=[]
                for member in dictionary['members']:
                    #print(member)
                    members.append(structAdjust(member))
                return members
        elif 'content' in dictionary.keys():
            if 'name' in dictionary.keys():
                content=structAdjust(dictionary['content'])
                return {dictionary['name']:content}
                
            else:
                return structAdjust(dictionary['content'])
        elif 'name' in dictionary.keys():
            return dictionary['name']
    elif type(dictionary)==list:
        members=[]
        for member in dictionary:
            members.append(structAdjust(member))
            #print(member)
            #print(recurse1(member))
            #print('\n')
        return members


def process(dictionary):
    for i in dictionary.keys():
        dictionary[i].pop('type')
        if 'value' in dictionary[i].keys():
            dictionary[i].pop('value')
        if 'named' in dictionary[i].keys():
            dictionary[i].pop('named')

    for (key,value) in dictionary.items():
        if 'content' in dictionary[key].keys():
            content=dictionary[key]['content']
            dictionary[key]=content
        elif 'members' in dictionary[key].keys():
            members=dictionary[key]['members']
            dictionary[key]=members
        #把只含有content或者members的字典的值提取出来
        getName(dictionary[key])
        #print(cpp_grammer_dict[key])#处理结果

    for (key,value)in dictionary.items():
        dictionary[key]=structAdjust(dictionary[key])
```
